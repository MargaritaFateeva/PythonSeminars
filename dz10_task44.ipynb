```python
import pandas as pd

```

```python
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

```python
data.loc[data['whoAmI']=='robot', 'robot']=1
data.loc[data['whoAmI']!='robot', 'robot']=0
data.loc[data['whoAmI']=='human', 'human']=1
data.loc[data['whoAmI']!='human', 'human']=0
data_res=data[['human', 'robot']]
data_res.head()
```

```python
data_1=pd.get_dummies(data['whoAmI'])
data_1.head()
```
